```python
from __future__ import division, unicode_literals

import os
import random
import pickle
import string
import nltk
import pandas as pd
import numpy as np
import math
from textblob import TextBlob
```

```python
path_arxiv = '/home/jerry/Data/Hopper_Project/ptm_data/arxiv_processed_trunc/'
path_wiki = '/home/jerry/Data/Hopper_Project/ptm_data/wiki/'

doclist_arxiv = {}
doclist_wiki = {}

vocab = set()

tfidf_arxiv = pd.DataFrame()
tfidf_wiki = pd.DataFrame()
```

```python
# alternate tfidf scores for dictionaries

def tf(word, blob):
    return blob.words.count(word) / len(blob.words)

def n_containing_dict(word, bloblist):
    return sum(1 for blob in bloblist if word in bloblist[blob]) # bc dictionary

def idf_dict(word, bloblist):
    return math.log(len(bloblist) / (1 + n_containing_dict(word, bloblist)))

def tfidf_dict(word, blob, bloblist):
    return tf(word, blob) * idf_dict(word, bloblist)
```

```python
# loop over all documents in wiki and arxiv folder
for subidr, dirs, files in os.walk(path_wiki):
    for file in files:
        file_path = path_wiki + file
        shakes = open(file_path, 'r')
        text = shakes.read()
        lowers = text.lower()
        no_punctuation = lowers.translate(None, string.punctuation)
        unicode_ignore = unicode(no_punctuation, errors='ignore')
        doclist_wiki[file] = TextBlob(unicode_ignore)

for subidr, dirs, files in os.walk(path_arxiv):
    for file in files:
        file_path = path_arxiv + file
        shakes = open(file_path, 'r')
        text = shakes.read()
        lowers = text.lower()
        no_punctuation = lowers.translate(None, string.punctuation)
        unicode_ignore = unicode(no_punctuation, errors='ignore')       
        #utf8_encoding = unicode(no_punctuation, 'utf-8')
        doclist_arxiv[file] = TextBlob(unicode_ignore)       
```

```python
# get the common vocab
# vocab is (initially) an empty set for the common vocabulary
# [12/17] we now make common vocab the intersection. There 
#         were too many words before, hopefully will speed
#         up computation time

vocab_wiki = set()
vocab_arxiv = set()

for i, doc in enumerate(doclist_wiki):
    word_set = set(doclist_wiki[doc].words)
    vocab_wiki = vocab_wiki.union(word_set)

for i, doc in enumerate(doclist_arxiv):
    word_set = set(doclist_arxiv[doc].words)
    vocab_arxiv = vocab_arxiv.union(word_set)
    
vocab = vocab_wiki.intersection(vocab_arxiv)
```

```python
# convert vocab to list, save as csv (for human-readable)
file_vocab_list = open('./intersect_vocab.txt', 'w')
for i in vocab:
    file_vocab_list.write('%s\n' % i)
    
# now we pickle it
pickle.dump(vocab, open('./intersect_vocab.p', 'w') )
```

```python
# take doclist_arxiv, and split into random subsets around size 100
# we hold these subset dictionaries in a list
doclist_arxiv_subsets = []

# make a copy of doclist_arxiv
doclist_arxiv_cp = doclist_arxiv.copy()

len_arxiv = len(doclist_arxiv)
n = 100
subset_size = 100
i = len_arxiv

def choose_subset(dictionary, size):
    # choose random keys
    keys = random.sample(dictionary, size)
    
    # get the list of associated keys
    values = [doclist_arxiv_cp[i] for i in keys]
    
    # create a new dictionary from these keys
    tmp_dict = dict( (keys[i], values[i]) for i in range(len(keys)) )
    
    # add dict subset to list
    doclist_arxiv_subsets.append(tmp_dict)
        
    # remove this subset from master
    for i in keys:
        del doclist_arxiv_cp[i] 

while i >= 0:
    if (i > n):
        subset_size = n
        choose_subset(doclist_arxiv_cp, subset_size)               
        
    else:
        subset_size = i
        choose_subset(doclist_arxiv_cp, subset_size)
        
    i -= n
```

```python
# for each document in doclist_wiki, doclist_arxiv, construct 
# a row in a pandas dataframe with N length vector for the 
# TFIDF score of each word in the common vocab vector in this
# document.

tfidf_wiki  = pd.DataFrame(columns = vocab)
tfidf_arxiv = pd.DataFrame(columns = vocab)

for i, doc in enumerate(doclist_wiki):
    scores = [tfidf_dict(word, doclist_wiki[doc], doclist_wiki) for word in vocab]
    tfidf_wiki.loc[doc] = scores
```

```python
# we don't need to compute tf-idf scores for all documents in 
# arxiv. Let's just do a subset of 100.

#for i in range(len(doclist_arxiv_subsets)):
#    print('doclist_arxiv_subsets[{}]'.format(i) )
#    doclist = doclist_arxiv_subsets[i]  

doclist = doclist_arxiv_subsets[0]
for j, doc in enumerate(doclist):
    scores = [tfidf_dict(word, doclist[doc], doclist) for word in vocab]
    tfidf_arxiv.loc[doc] = scores 
```

```python
# normalization function
def magnitude(v):
    return math.sqrt(sum(i*i for i in v))

def normalize(v):
    vmag = magnitude(v)
    if vmag == 0:
        print('vmag is zero')
        return 0
    else:
        return [ i/vmag for i in v ]
    
def dot_product(u,v):
    return sum(u[i]*v[i] for i in range(len(u)))
```

```python
# let's write a function to take in a arxiv paper name and output the top k wiki concepts

def relevant_concepts(arxiv_doc, doc_comparisons):
    print("\nTop words in document: {}".format(arxiv_doc))
    
    # check if arxiv_doc in doc_comparsions
    if arxiv_doc in doc_comparisons.keys(): 
        scores = doc_comparisons[arxiv_doc]
        sorted_words = sorted(scores.items(), key=lambda x : x[1], reverse = True)
        for concept, score in sorted_words[:10]:
            print("\tConcept: {}, TF-IDF: {}".format(concept, round(score, 5)))
    
    else:
        print("\tError: input doc not in doc_comparisons")
        
def show_doc_text(arxiv_doc):
    f = open('/home/jerry/Data/Hopper_Project/ptm_data/arxiv_processed_trunc/' + arxiv_doc, 'r')
    text = f.read()
    f.close()
    print('\n' + arxiv_doc + '\n\n' + text)
```
