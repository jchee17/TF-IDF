```python
from __future__ import division, unicode_literals

import os
import random
import pickle
import string
import nltk
import math
from textblob import TextBlob
```

```python
# Load pickled items

vocab = pickle.load(open('../data_objects/vocab.p', 'r'))

wiki_sep_byword = pickle.load(open('./data_objects/wiki_sep_byword.p', 'r'))
arxiv_sep_byword = pickle.load(open('./data_objects/arxiv_sep_byword.p', 'r'))

wiki_sep_byfile = pickle.load(open('./data_objects/wiki_sep_byfile.p', 'r'))
arxiv_sep_byfile = pickle.load(open('./data_objects/arxiv_sep_byfile.p', 'r'))

wiki_wordsindoc = pickle.load(open('./data_objects/wiki_wordsindoc.p', 'r'))
arxiv_wordsindoc = pickle.load(open('./data_objects/arxiv_wordsindoc.p', 'r'))
```

```python
# code to turn all values from string to int
tmp = None

for f in wiki_sep_byword:
    tmp = int(wiki_sep_byword[f])
    wiki_sep_byword[f] = tmp
    
for f in arxiv_sep_byword:
    tmp = int(arxiv_sep_byword[f])
    arxiv_sep_byword[f] = tmp
    
for f in wiki_sep_byfile:
    for key in wiki_sep_byfile[f]:
        tmp = int(wiki_sep_byfile[f][key])
        wiki_sep_byfile[f][key] = tmp
        
for f in arxiv_sep_byfile:
    for key in arxiv_sep_byfile[f]:
        tmp = int(arxiv_sep_byfile[f][key])
        arxiv_sep_byfile[f][key] = tmp
        
for f in wiki_wordsindoc:
    tmp = int(wiki_wordsindoc[f])
    wiki_wordsindoc[f] = tmp
    
for f in arxiv_wordsindoc:
    tmp = int(arxiv_wordsindoc[f])
    arxiv_wordsindoc[f] = tmp
```

```python
# pickle everything
pickle.dump(wiki_sep_byword, open('./data_objects/wiki_sep_byword.p', 'w'))
pickle.dump(arxiv_sep_byword, open('./data_objects/arxiv_sep_byword.p', 'w'))

pickle.dump(wiki_sep_byfile, open('./data_objects/wiki_sep_byfile.p', 'w'))
pickle.dump(arxiv_sep_byfile, open('./data_objects/arxiv_sep_byfile.p', 'w'))

pickle.dump(wiki_wordsindoc, open('./data_objects/wiki_wordsindoc.p', 'w'))
pickle.dump(arxiv_wordsindoc, open('./data_objects/arxiv_wordsindoc.p', 'w'))
```

```python
# functions for tf-idf score
length_wiki = wiki_sep_byword['number of documents in wiki']
length_arxiv = arxiv_sep_byword['number of documents in arxiv']

def tf(word, document, corpus):
    # check string input
    if (corpus == 'wiki'):
        return (wiki_sep_byfile[document][word] / wiki_wordsindoc[document])
    elif (corpus == 'arxiv'):
        return (arxiv_sep_byfile[document][word] / arxiv_wordsindoc[document])
    else:
        return None

def idf(word, corpus):
    # check string input
    if (corpus == 'wiki'):
        return math.log(length_wiki / (1 + wiki_sep_byword[word]))
    elif (corpus == 'arxiv'):
        return math.log(length_arxiv/ (1 + arxiv_sep_byword[word]))
    else:
        return None
    
# note: corpus a string designation
def tfidf(word, document, corpus):
    return tf(word, document, corpus) * idf(word, corpus)
```

```python
# compute tf-idf scores
tfidf_wiki = {}
tfidf_arxiv = {}

for f in wiki_sep_byfile:
    #print(f)
    scores = [tfidf(word, f, 'wiki') for word in vocab]
    tfidf_wiki[f] = scores
    
for f in arxiv_sep_byfile:
    scores = [tfidf(word, f, 'arxiv') for word in vocab]
    tfidf_arxiv[f] = scores
```

```python
pickle.dump(tfidf_wiki, open('./data_objects/tfidf_wiki.p', 'w'))
pickle.dump(tfidf_arxiv, open('./data_objects/tfidf_arxiv.p', 'w'))
```

```python
# normalization function
def magnitude(v):
    return math.sqrt(sum(i*i for i in v))

def normalize(v):
    vmag = magnitude(v)
    if vmag == 0:
        print('vmag is zero')
        return 0
    else:
        return [ i/vmag for i in v ]
    
def dot_product(u,v):
    return sum(u[i]*v[i] for i in range(len(u)))
```

```python
# now let's normalize the vectors in tfidf_wiki, tfidf_arxiv

norm = []

tfidf_wiki_norm = {}
tfidf_arxiv_norm = {}

for f in tfidf_wiki:
    norm = normalize(tfidf_wiki[f])
    #print(index, magnitude(norm))
    tfidf_wiki_norm[f] = norm

for f in tfidf_arxiv:
    norm = normalize(tfidf_arxiv[f])
    tfidf_arxiv_norm[f] = norm
    #print(index, magnitude(norm))
    
pickle.dump(tfidf_wiki_norm, open('./data_objects/tfidf_wiki_norm.p', 'w'))
pickle.dump(tfidf_arxiv_norm, open('./data_objects/tfidf_arxiv_norm.p', 'w'))
```

```python
# document comparisons
doc_comparisons = {}

for f in tfidf_arxiv_norm:
    doc_comparisons[f] = {key:dot_product(tfidf_arxiv_norm[f], tfidf_wiki_norm[key]) 
                            for key in tfidf_wiki_norm}

pickle.dump(doc_comparisons, open('./data_objects/doc_comparisons.p', 'w'))
```
