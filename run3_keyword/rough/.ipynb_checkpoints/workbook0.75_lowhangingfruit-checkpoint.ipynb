```python
import os
import re
import codecs
import timeit
import string
import pickle

path_arxiv = '/home/jerry/Data/Hopper_Project/ptm_data/arxiv_processed_trunc/'
articles = os.listdir(path_arxiv)

concepts = pickle.load(open('./data_objects/u_master-concepts.p', 'r'))

#arxiv_sep_byfile = pickle.load(open('./data_objects/arxiv_sep_byfile.p' ,'r'))
#arxiv_wordsindoc = pickle.load(open('./data_objects/arxiv_wordsindoc.p', 'r'))
```

```python
# check so that there are no spaces at beginning or end of 
# keywords in concepts
for word in concepts:
    if (' ' == word[-1:]):
        print(word)
        
    if ('' == word[:1]):
        print(word)
        
# check so that there are no '[]' in concepts
for word in concepts:
    if ('[]' in word):
        print(word)
```

```python
# I try to use re.sub to replace [-\– ] with the 
# literal [-\– ] to be used in a regex findall
# NOTE: if I try to add word boundaries ur'\b' I get fewer matches

re_concepts = {}

def create_regex(keyword):
    re_keyword = ur''
    re_keyword = re.sub(ur"(\\-)|(\\–)|(\\\s)", u"[-\s–]", re.escape(keyword))
    #re_keyword = ur'\b' + re.sub(ur'[-\– ]', ur'[-\– ]', re.escape(keyword)) + ur'\b'
    re_concepts[keyword] = re_keyword
    #print(re_keyword)
    
for word in concepts:
    create_regex(word)
    
re_concepts.pop('')
```

```python
for key, r in re_concepts.iteritems():
    print(key)
    print(r)
```

```python
fp = None
txt = u''
#txt = ""
arxiv_sep_byfile_re = {}
#arxiv_sep_byfile = {}
num_r = 0
start = timeit.default_timer()
elapsed = 0

count = 0
for f in articles:
    # import text
    fp = codecs.open(path_arxiv+f, "r", "utf-8", errors='ignore')
    #fp = open(path_arxiv+f, 'r')
    txt = fp.read()
    txt = txt.lower()
    
    # create sub dictionary
    arxiv_sep_byfile_re[f] = {}
    #arxiv_sep_byfile[f] = {}
    
    # let's call re.findall
    for keyword, r in re_concepts.iteritems():
        #print(r)
        num_r = len(re.findall(r, txt, re.UNICODE))
        
        if (num_r > 0):
            arxiv_sep_byfile_re[f][keyword] = num_r
            
    if (count % 100 == 0): 
        elapsed = timeit.default_timer() - start
        start = timeit.default_timer()
        print(count, elapsed)
    count+=1
    
    
```

```python
arxiv_sep_byword_re = {}

for keyword, r in re_concepts.iteritems():
    arxiv_sep_byword_re[keyword] = 0
    
for f in arxiv_sep_byfile_re:
    for keyword in arxiv_sep_byfile_re[f]:
        arxiv_sep_byword_re[keyword] += 1
        
pickle.dump(arxiv_sep_byword_re, open('./data_objects/arxiv_sep_byword_re.p', 'w'))
```
